```python
import yaml
import os
from PIL import Image
```

```python
def txtread(path):
    path = os.path.expanduser(path)
    with open(path, 'r') as f:
        return f.read()
```

```python
def yamlread(path):
    return yaml.safe_load(txtread(path=path))
```

```python
def imwrite(path=None, img=None):
    Image.fromarray(img).save(path)
```
